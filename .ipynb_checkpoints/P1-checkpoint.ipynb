## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    
    ## Equation for line y = mx + b, get all parameters for left line and right line
    
    # Step 1: Bin all right line parameters and left line parameters
    
    # Bins for Right Lane parameters
    right_linex = []
    right_liney = []
    pos_slopes = []
    pos_intercepts = []
    
    # Bins for Left Lane parameters
    left_linex = []
    left_liney = []
    neg_slopes = []
    neg_intercepts = []

    ymax = img.shape[0]
    ymin = img.shape[0] 
    
    for line in lines:
        for x1,y1,x2,y2 in line:
            gradient = (y2-y1)/(x2-x1)
            
            if gradient > 0.50 and gradient < math.inf:
                pos_slopes.append(gradient)  # Bin all parameters for right lane
                right_linex.append(x1)
                right_linex.append(x2) 
                right_liney.append(y1)
                right_liney.append(y2)
                pos_intercepts.append(y1-gradient*x1)
                
            if gradient < -0.50 and gradient > -math.inf:
                neg_slopes.append(gradient)  # Bin all parameters for left lane
                left_linex.append(x1)
                left_linex.append(x2) 
                left_liney.append(y1)
                left_liney.append(y2)
                neg_intercepts.append(y1-gradient*x1)
                
        ymin = min(ymin,y1,y2)
        
    ymin += -10 # Adjust ymin to control horizon cross over point for right and left lines
            
    # Step 2: Average all line positions, use x = (y-c)/m to find corresponding x parameter for given y and slope
    
    # Average all right line positions
    if len(pos_slopes)>0 and len(right_linex)>0 and len(right_liney)>0:
        avg_pos_slope = sum(pos_slopes)/len(pos_slopes)
        ave_right_linex = sum(right_linex)/len(right_linex)
        ave_right_liney = sum(right_liney)/len(right_liney)
        ave_pos_intercept = sum(pos_intercepts)/len(pos_intercepts)
        xmin = int((ymin-ave_pos_intercept)/avg_pos_slope)
        xmax = int((ymax-ave_pos_intercept)/avg_pos_slope)            
        cv2.line(img, (xmin, ymin), (xmax, ymax), color, 12)
            
    # Average all left line positions
    if len(neg_slopes)>0 and len(left_linex)>0 and len(left_liney)>0:
        avg_neg_slope = sum(neg_slopes)/len(neg_slopes)
        ave_left_linex = sum(left_linex)/len(left_linex)
        ave_left_liney = sum(left_liney)/len(left_liney)
        ave_neg_intercept = sum(neg_intercepts)/len(neg_intercepts)
        xmin = int((ymin-ave_neg_intercept)/avg_neg_slope)
        xmax = int((ymax-ave_neg_intercept)/avg_neg_slope)            
        cv2.line(img, (xmin, ymin), (xmax, ymax), color, 12)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform, returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
````

````python
# TODO: Build your pipeline that will draw lane lines on the test_images
# then save them to the test_images directory.


######## MODIFY THESE VARIABLES TO MAKE YOUR COLOR SELECTION ##########

# Set image kernel size
ksize = 5       

# Define the vertices of a mask. Image size is (540,960) with 3 numbers (RGB)for each element
imgx = image.shape[1] # This is 960, Xmax
imgy = image.shape[0] # This is 540, Ymax

# The area of interest is defined as ratio of the image
left_bottomx = 0.085*imgx  
right_bottomx = 0.92*imgx
apexy = 0.59*imgy     # Height of the apex y coordinate
d = 40           # Distance between the two apex points making rectangle

# Threshold for Canny Edge Detection gradient
low_t = 60      
high_t = 180 

# Parameters for Hough Transform Function def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
rho = 3                # distance resolution in pixels of the Hough grid
theta = np.pi/180      # angular resolution in radians of the Hough grid
threshold = 100        # minimum number of votes (intersections in Hough grid cell)
min_line_len = 1       #minimum number of pixels making up a line
max_line_gap = 1000     # maximum gap in pixels between connectable line segments


def lanedetect_pipe(image):
    
    # Pipeline building occurs in following steps
    
    # Step 1: Convert to gray 
    gray_image = grayscale(image)
    
    # Step 2: Gaussian blurring of image before edge detection
    smooth_image = gaussian_blur(gray_image, ksize)  
    
    # Step 3: Canny edge detection
    edge_image = canny(smooth_image, low_t, high_t)   
    
    # Step 4: Define the area of interest for lane detection
    quad_vertices = np.array([[(left_bottomx,imgy),(imgx/2-d/2,apexy),(imgx/2+d/2,apexy),
                          (right_bottomx,imgy)]],dtype=np.int32)
    
    quad = region_of_interest(edge_image,quad_vertices)
    
    # Step 5: Perform Hough transform for identification of line segments in the region of interest
    hough_img = hough_lines(quad, rho, theta, threshold, min_line_len, max_line_gap)
    
    # Step 6: Merge line image with original image
    merge_image = weighted_img(hough_img,image)
    
    return merge_image

for test_image in os.listdir("test_images/"):
    image = mpimg.imread('test_images/' + test_image)
    image1 = gaussian_blur(grayscale(image), 5)  
    image2 = canny(image1,60,180)
    quad_vertices = np.array([[(left_bottomx,imgy),(imgx/2-d/2,apexy),(imgx/2+d/2,apexy),
                          (right_bottomx,imgy)]],dtype=np.int32)
    
    quad = region_of_interest(image2,quad_vertices)
    
    image3 = hough_lines(quad, rho, theta, threshold, min_line_len, max_line_gap)
    
    image4 = weighted_img(image3,image)
    plt.imshow(image3)
    plt.imshow(lanedetect_pipe(image))  # if you wanted to show a single color channel image called 'gray', for example, call as plt.imshow(gray, cmap='gray')
    plt.show()
````
